**Importing the modules**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

**Reading the image**

```python
img = cv2.imread('/content/sample.jpg', cv2.IMREAD_UNCHANGED)
```

**Showing the image**

```python
cv2_imshow(img)
```

**Converting in gray scale image**

```python
img=cv2.imread("/content/sample.jpg",0)
```

**Writing the image**

```python
cv2.imwrite('New_sample.jpg',img)
```

```python
cv2_imshow(img)
```

**Inverting the image**

```python
inverted_image=255-img
('Inverted',inverted_image)
cv2.waitKey()
```

**Blurring the image**

```python
blurred=cv2.GaussianBlur(inverted_image,(21,21),0)
gray_image= ('New_sample.jpg',img)
cv2_imshow(blurred)
```

```python
inverted_blurred=255-blurred
```

```python
dodge=lambda image,mask:cv2.divide(image,255-mask,scale=256)
blanded=dodge(img,blurred)
```

```python
plt.imshow(blanded,cmap='gray')
```
